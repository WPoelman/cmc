```python
import pandas as pd

df = pd.read_csv('data/cmc_data_export_with_coding_results.csv')
```

```python
df.head()
```

```python
col_to_str = {
    "timestamp": "Tijdstempel",
    "work_tasks": "1. Wat zijn taken die jij regelmatig uitvoert voor je werk? Probeer specifiek te zijn en minstens 5 te noemen. (Denk bijvoorbeeld aan tentamencijfers invoeren in ProgressWWW.)",
    "work_systems": "2. Met welke systemen werk je zoal?",
    "face_to_face": "3. Hoe goed vind je dat je uitleg kan geven in een face-to-face gesprek?",
    "google_meet_no_screen": "4.1 Hoe goed vind je dat je uitleg kan geven in een gesprek via Google Meet?",
    "google_meet_screen": "4.2 Hoe goed vind je dat je uitleg kan geven via Google Meet wanneer je je scherm moet delen?",
    "google_hangouts": "5. Hoe goed vind je dat je uitleg kan geven via Google Hangouts?",
    "whatsapp": "6. Hoe goed vind je dat je uitleg kan geven via Whatsapp?",
    "email": "7. Hoe goed vind je dat je uitleg kan geven via E-mail?",
    "slack": "8. Hoe goed vind je dat je uitleg kan geven via Slack?",
    "location_preference": "9. Werk je liever thuis of op kantoor?",
    "location_preference_expl": "10. Zou je je antwoord op vraag 9 kunnen toelichten?",
}
str_to_col = {v: k for k, v in col_to_str.items()}
```

```python
df.columns = df.columns.map(lambda c: str_to_col[c] if c in str_to_col else c)
```

```python
df
```

```python
def get_score(row, items):
    # 0 means: "I don't use this tool"
    applicable_items = [row[item] for item in items if row[item] != 0]

    return sum(applicable_items) / len(applicable_items)


competency_items = [
    "face_to_face",
    "google_meet_no_screen",
    "google_meet_screen",
    "google_hangouts",
    "whatsapp",
    "email",
    "slack",
]

df["overall_competence_score"] = df.apply(
    lambda row: get_score(row, competency_items), 
    axis=1
)
```

```python
online_competency_items = [
    "google_meet_no_screen",
    "google_meet_screen",
    "google_hangouts",
    "whatsapp",
    "email",
    "slack",
]
df["online_competence_score"] = df.apply(
    lambda row: get_score(row, online_competency_items), 
    axis=1
)
```

```python
print('Average score per communication method')
for item in competency_items:
    selection = df[item] != 0
    print(f'{item}: {round(df[selection][item].mean(), 2)} (used by {len(df[selection])})')
```

```python
edge_records, node_records = [], []
for idx, row in df.iterrows():
    node_records.append({
        "Id": idx, 
        "Total Competence": round(row["overall_competence_score"], 2),
        "Online Competence": round(row["online_competence_score"], 2),
        "F2F Competence": row["face_to_face"],
        "Location Preference": row["location_preference"],
    })

    for tool in online_competency_items:
        if row[tool] != 0:
            edge_records.append(
                {
                    "Source": idx,
                    "Target": tool,
                    "Type": "Undirected",
                    "Weight": row[tool],
                }
            )


pd.DataFrame().from_records(edge_records).to_csv(
    "gephi_test_tools_only_edges.csv", index=False
)

pd.DataFrame().from_records(node_records).to_csv(
    "gephi_test_tools_only_nodes.csv", index=False
)
```

```python
knowledge_areas = [
    'aanpassingen_nestor',
    'afstudeerprocedure',
    'cijferverwerking',
    'contractuele_zaken',
    'dataverwerking_ocasys',
    'diplomas_afgeven',
    'examencommissie_besluiten',
    'kwaliteitswaarborging_onderwijs',
    'niet_reguliere_administratie',
    'ontvangst_exchange_students',
    'roosteren',
    'stagecontracten',
    'vakintekeningen',
    'verwerking_exchange_students',
    'zaalaanvragen',
]
```

```python
records = []
for knowledge_area in knowledge_areas:
    face_to_face = round(df[df[knowledge_area].notnull()]['face_to_face'].mean(), 2)
    online_competence_score = round(df[df[knowledge_area].notnull()]['online_competence_score'].mean(), 2)
    overall_competence_score = round(df[df[knowledge_area].notnull()]['overall_competence_score'].mean(), 2)
    records.append({
        'knowledge_area': knowledge_area,
        'face_to_face': face_to_face,
        'online_competence_score': online_competence_score,
        'overall_competence_score': overall_competence_score,
    })
df_new = pd.DataFrame().from_records(records)
```
