```python
col_to_str = {
    "timestamp": "Tijdstempel",
    "work_tasks": "1. Wat zijn taken die jij regelmatig uitvoert voor je werk? Probeer specifiek te zijn en minstens 5 te noemen. (Denk bijvoorbeeld aan tentamencijfers invoeren in ProgressWWW.)",
    "work_systems": "2. Met welke systemen werk je zoal?",
    "face_to_face": "3. Hoe goed vind je dat je uitleg kan geven in een face-to-face gesprek?",
    "google_meet_no_screen": "4.1 Hoe goed vind je dat je uitleg kan geven in een gesprek via Google Meet?",
    "google_meet_screen": "4.2 Hoe goed vind je dat je uitleg kan geven via Google Meet wanneer je je scherm moet delen?",
    "google_hangouts": "5. Hoe goed vind je dat je uitleg kan geven via Google Hangouts?",
    "whatsapp": "6. Hoe goed vind je dat je uitleg kan geven via Whatsapp?",
    "email": "7. Hoe goed vind je dat je uitleg kan geven via E-mail?",
    "slack": "8. Hoe goed vind je dat je uitleg kan geven via Slack?",
    "location_preference": "9. Werk je liever thuis of op kantoor?",
    "location_preference_expl": "10. Zou je je antwoord op vraag 9 kunnen toelichten?",
}
str_to_col = {v: k for k, v in col_to_str.items()}
```

```python
df.columns = df.columns.map(lambda c: str_to_col[c])
```

```python
df
```

```python
def get_score(row):
    competency_items = [
        "google_meet_no_screen",
        "google_meet_screen",
        "google_hangouts",
        "whatsapp",
        "email",
        "slack",
    ]
    # 0 means: "I don't use this tool"
    applicable_items = [row[item] for item in competency_items if row[item] != 0]

    return sum(applicable_items) / len(applicable_items)


df["competence_score"] = df.apply(get_score, axis=1)
```

```python
df[['face_to_face', 'competence_score', 'location_preference', 'location_preference_expl']]
```

```python
competency_items = [
    "face_to_face",
    "google_meet_no_screen",
    "google_meet_screen",
    "google_hangouts",
    "whatsapp",
    "email",
    "slack",
]

print('Average score per communication method')
for item in competency_items:
    selection = df[item] != 0
    print(f'{item}: {round(df[selection][item].mean(), 2)} (used by {len(df[selection])})')
```

```python
online_communication_tools = [
    "google_meet_no_screen",
    "google_meet_screen",
    "google_hangouts",
    "whatsapp",
    "email",
    "slack",
]

edge_records, node_records = [], []
for idx, row in df.iterrows():
    node_records.append({
        "Id": idx, 
        "Total Competence": round(row["competence_score"], 2),
        "F2F Competence": row["face_to_face"],
        "Location Preference": row["location_preference"],
    })

    for tool in online_communication_tools:
        if row[tool] != 0:
            edge_records.append(
                {
                    "Source": idx,
                    "Target": tool,
                    "Type": "Undirected",
                    "Weight": row[tool],
                }
            )


pd.DataFrame().from_records(edge_records).to_csv(
    "gephi_test_tools_only_edges.csv", index=False
)

pd.DataFrame().from_records(node_records).to_csv(
    "gephi_test_tools_only_nodes.csv", index=False
)
```
